```python
import pandas as pd
import seaborn as sns
import pingouin as pg 
from scipy.stats import chi2_contingency
```

```python
adults = pd.read_csv('csv_data/adult.csv')
```

Вопрос:

Связан ли уровень образования (education) с тем, зарабатывает ли человек больше 50K (income)?

```python
adults.head()
```

Так как 2 переменные категориальные, то для проверки статистической значимости используем ХИ квадрат. Проверим сразу на двух библиотеках.

```python
table = pd.crosstab(adults['education'], adults['income'])
table
```

Тест здесь применим, так как обе переменные категориальные. Однако наличие редких категорий может нарушать предпосылки теста (малые ожидаемые частоты), поэтому целесообразно либо объединить категории, либо интерпретировать результаты с осторожностью.

```python
chi2_contingency(table)
```

```python
pg.chi2_independence(adults, x = 'education', y = 'income')
```

Вывод: Существует статистически значимая связь между образованием и вероятностью получить определенный уровень дохода (pvalue=0.0), умеренная практическая значимость (cramer  = 0.368)
